#### Finding The Peak Hours

### Recommender

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(pizza_name, df = df):
    df_rec = df.copy()
    
    text_data = df_rec['pizza_category'] + df_rec['pizza_ingredients'] + df_rec['pizza_name']
    rel_features = ['pizza_name', 'pizza_ingredients', 'pizza_category']
    
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(text_data)
    
    query_vec = tfidf.transform([pizza_name])
    query_sim = cosine_similarity(query_vec, tfidf_matrix)[0]
    query_rec = query_sim.argsort()[::-1][:1]
    
    return df_rec[rel_features].iloc[query_rec]
```

```python
recommend("Veggie spinach")
```
